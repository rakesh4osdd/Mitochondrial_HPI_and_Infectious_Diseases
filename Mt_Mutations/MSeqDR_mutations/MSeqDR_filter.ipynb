```python
# mapping MSeqDR trasncript to mutation and diseases
import pandas as pd
pd.__version__
```

```python
# read transcript
ncbi_id= pd.read_csv("NCBI_Protein_ID_uniprot_ids_5778_tr_ids1788_c1_1780.txt", sep='\t', names=["NCBI Protein ID"])
print(ncbi_id.shape)
ncbi_id.head()
```

```python
# read all variants
variants= pd.read_csv("MSeqDR_all_variants_transcripts_data.txt", sep='\t', low_memory=False)
print(variants.shape)
variants.head()
```

```python
# inner join on 'NCBI Protein ID'
variants_ncbi_id= pd.merge(ncbi_id, variants, on='NCBI Protein ID')
variants_ncbi_id.to_csv("variants18976_ncbi_id1780.tsv", sep="\t", index=False)
print(variants_ncbi_id.shape)
variants_ncbi_id.head()
```

```python
#find unique genes
variants_ncbi_id.Gene.unique()
variants_ncbi_id.Gene.nunique()
variants_genes = pd.DataFrame(variants_ncbi_id.Gene.unique())
variants_genes.columns=['Symbol']
print(variants_genes.shape)
variants_genes.head()
```

```python
genes= pd.read_csv("MSeqDR_all_genes_to_disease.txt", sep='\t', low_memory=False)
```

```python
print(genes.shape)
genes.head()
```

```python
# get mapped genes to get list of diseases
# inner join on 'Symbol'
genes_variants_ncbi_id= pd.merge(variants_genes, genes, on='Symbol')
genes_variants_ncbi_id.to_csv("gene284_variants18976_ncbi_id1780.tsv", sep="\t", index=False)
print(genes_variants_ncbi_id.shape)
genes_variants_ncbi_id.head()
```
